```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
x1=np.linspace(0,10,30) ## Number of dots in plot
fig=plt.figure()
plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
plt.plot(x1,np.cos(x1),'*')
plt.show()
```

```python
plt.subplot(2,1,1) ##(rows,columns,panel no.)
plt.plot(x1,np.sin(x1))
```
